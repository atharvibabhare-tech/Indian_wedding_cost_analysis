```python
import ipywidgets as widgets
from ipywidgets import interact
```

```python
def dashboard(Wedding_Type, Place, Decor_Category):
    
    filtered_df = df[
        (df['Wedding_Type'] == Wedding_Type) &
        (df['Place'] == Place) &
        (df['Decor_Category'] == Decor_Category)
    ]
    
    result = filtered_df.groupby('Entertainment_Category')['Entertainment'].mean()
    plt.xticks(rotation=45)
    result.plot(kind='bar', color='orange', edgecolor='black')
    plt.show()
```

```python
interact(
    dashboard,
    Wedding_Type=df['Wedding_Type'].unique(),
    Place=df['Place'].unique(),
    Decor_Category=df['Decor_Category'].unique()
)
```

### Interactive Dashboard

This dashboard allows users to filter data based on:
- Wedding Type
- Place
- Decor Category

After applying filters, it shows the average entertainment cost by category.

Users can interactively explore how entertainment costs change based on different conditions.

Insight:
This helps in identifying cost patterns dynamically instead of static analysis.

## Insights from Pairplot

- A pairplot shows **relationships between all numerical variables** in one view.
- Diagonal graphs show the **distribution (spread) of each variable**.
- Off-diagonal graphs (scatter plots) show **how two variables are related**.

### Observations

- Most scatter plots show **random distribution of points**.
- This means there is **no strong correlation** between many variables.
- Some plots may show a **slight trend**, but not very strong.
- The diagonal plots show that most values are **concentrated in mid-range**.

```python
sb.jointplot(x='Photography', y='Cost_of_Type', data=df, kind='scatter')
plt.show()
```

## Insights from Jointplot (Photography vs Cost of Type)
- There is **no strong relationship** between photography cost and cost of type.
- The points are **randomly scattered**, meaning one does not directly affect the other.
- Photography costs are spread between **₹50,000 to ₹3,00,000**.
- Cost of type ranges widely, mostly between **₹2,00,000 to ₹8,00,000+**.
- A dense cluster is visible in the middle, showing most data lies in **mid-range values**.
- Very high values of cost of type are **less frequent**.

```python
sb.jointplot(x='Decor', y='Cost_of_Type', data=df, kind='scatter')
plt.show()
```

## Insights from Jointplot (Decor vs Cost of Type)
- There is **no clear relationship** between decor cost and cost of type.
- The data points are **randomly scattered**, showing both variables are independent.
- Decor costs range from **₹50,000 to ₹6,00,000**.
